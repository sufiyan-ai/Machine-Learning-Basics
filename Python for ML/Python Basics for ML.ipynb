### Filtering

```python
# find out all rides greater than 10 miles
df.loc[df["MILES*"] > 10].head()
```

```python
df2 = df.loc[df["MILES*"] > 10,["MILES*",'START_DATE*']]
df2.head()
```

```python
df3 = df2.iloc[0:3]
df3.head()
```

```python
# Find out all rides to Cary
df.loc[df["START*"]=="Cary",["MILES*"]].head()
```

```python
# Find out all rides from Cary & Morrisville          
df.loc[df["START*"].isin(["Cary","Morrisville"])].shape
```

```python
# Find out all rides to Cary & Morrisville          
df[df["STOP*"].isin(["Cary","Morrisville"])].shape
```

```python
df.loc[~df["START*"].isin(["Cary","Morrisville"])].shape
```

```python
#extracting top 3 rows that are starting from cary and Morrisville with more than 10 miles
df4=df.loc[(df['MILES*']>10) & (df['START*'].isin(['Cary','Morrisville']))]
df4.iloc[:3,:]
```

### Groupby Summarize

```python
# for each start loc find the mean distance travelled
df.groupby("START*")["MILES*"].agg("mean")
```

## calculating with new data

```python
store_sales = pd.read_csv("store sales.csv")
```

```python
store_sales.head()
```

```python
def bonus_func(sales_col,sales_num):
    eligible_values=np.where(sales_col>sales_num,'Eligible','Not Eligible')
    return eligible_values
store_sales['Bonus']=bonus_func(store_sales['Jan'],10)
store_sales
```

```python
def bonus_func(sales_col):
    eligible_values=np.where(sales_col>10,'Eligible','Not Eligible')
    return eligible_values

allmon=store_sales.select_dtypes(include=['int64','float64']).apply(bonus_func)
allmon
```

```python
df=pd.concat([store_sales,allmon],axis=1)
df
```
